## Iteration:
萬物皆物件，萬物皆可loop (iterable)

```python
# Bad one: all in python are objeects & iterable
my_list = range(10)
i = 0
while i < len(my_list):
    v = my_list[i]
    print (v)
    i += 1
```

```python
# better one:
my_list = range(10)
for i in range(len(my_list)):
    v = my_list[i]
    print (v)
```

```python
# pythonic
for v in my_list:
    print (v)
```

## Q: How do I get the index?

```python
#NG
my_list = range(10,0,-1)
for i in range(len(my_list)):
    v = my_list[i]
    print (i, v)
```

```python
# good
my_list = range(10,0,-1)
for i, v in enumerate(my_list):
    print (i, v)
```
